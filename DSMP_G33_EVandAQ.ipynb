# Finding the correlation between EVs and Air quality

### Import Neccessary library

```python
# Import necessary library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Step1 - Import data and clean the data

```python
# Read .csv file and then turn into dataframe
df_air_quality = pd.read_csv("C:/Users/Admin/Desktop/DSMP/Raw data/air-quality-data-continuous.csv", sep=";")
df_EV_pop = pd.read_csv("C:/Users/Admin/Desktop/DSMP/Raw data/Bristol car sales data -EV population.csv")
```

```python
df_EV_pop
```

```python
# Filter the data to only include rows where Fuel is 'Battery electric', ONS Geography is 'Bristol, City of', and Keepership is 'Total'
battery_electric = df_EV_pop[(df_EV_pop['Fuel'] == 'Battery electric') & 
                             (df_EV_pop['ONS Geography [note 6]'] == '   Bristol, City of') & 
                             (df_EV_pop['Keepership'] == 'Total')]

# remove commas from the numbers
battery_electric = battery_electric.replace(',', '', regex=True)

# Print table
battery_electric
```

```python
# Create a list of the quarters in the data set
quarters = ['2012 Q1', '2012 Q2', '2012 Q3', '2012 Q4', 
            '2013 Q1', '2013 Q2', '2013 Q3', '2013 Q4', 
            '2014 Q1', '2014 Q2', '2014 Q3', '2014 Q4', 
            '2015 Q1', '2015 Q2', '2015 Q3', '2015 Q4', 
            '2016 Q1', '2016 Q2', '2016 Q3', '2016 Q4', 
            '2017 Q1', '2017 Q2', '2017 Q3', '2017 Q4', 
            '2018 Q1', '2018 Q2', '2018 Q3', '2018 Q4', 
            '2019 Q1', '2019 Q2', '2019 Q3', '2019 Q4', 
            '2020 Q1', '2020 Q2', '2020 Q3', '2020 Q4', 
            '2021 Q1', '2021 Q2', '2021 Q3', '2021 Q4', 
            '2022 Q1', '2022 Q2', '2022 Q3']
quarters_Q1 = ['2012 Q1', 
            '2013 Q1',
            '2014 Q1',
            '2015 Q1', 
            '2016 Q1',
            '2017 Q1', 
            '2018 Q1', 
            '2019 Q1',
            '2020 Q1', 
            '2021 Q1', 
            '2022 Q1']

quarters_Q2 = ['2012 Q2', 
               '2013 Q2',
               '2014 Q2',
               '2015 Q2',
               '2016 Q2',
               '2017 Q2',
               '2018 Q2',
               '2019 Q2',
               '2020 Q2',
               '2021 Q2',
               '2022 Q2']

quarters_Q3 = ['2012 Q3',
               '2013 Q3',
               '2014 Q3',
               '2015 Q3',
               '2016 Q3',
               '2017 Q3',
               '2018 Q3',
               '2019 Q3',
               '2020 Q3',
               '2021 Q3',
               '2022 Q3']

quarters_Q4 = ['2012 Q4',
               '2013 Q4',
               '2014 Q4', 
               '2015 Q4', 
               '2016 Q4', 
               '2017 Q4', 
               '2018 Q4', 
               '2019 Q4', 
               '2020 Q4', 
               '2021 Q4']

# Create a list of the corresponding values for each quarter in the Battery electric data set
values = battery_electric[quarters].values.tolist()[0]
values_Q1 = battery_electric[quarters_Q1].values.tolist()[0]
values_Q2 = battery_electric[quarters_Q2].values.tolist()[0]
values_Q3 = battery_electric[quarters_Q3].values.tolist()[0]
values_Q4 = battery_electric[quarters_Q4].values.tolist()[0]
```

```python
# Resample the data to quarterly frequency and calculate the mean of NOx for each quarter
df_quarterly = df_air_quality['NOx'].resample('Q').mean()

# Convert the DatetimeIndex to a PeriodIndex and rename it to show the year and quarter
df_quarterly.index = df_quarterly.index.to_period('Q').strftime('%Y Q%q')

# Print the quarterly mean of NOx
pd.DataFrame(df_quarterly)
```

```python
# Create dataframe of EVs before merge with NOx data frame
df_battery_electric = pd.DataFrame(list(zip(quarters, values)), columns=['Quarters', 'Number of Vehicles'])
df_battery_electric_Q1 = pd.DataFrame(list(zip(quarters_Q1, values_Q1)), columns=['Quarters', 'Number of Vehicles'])
df_battery_electric_Q2 = pd.DataFrame(list(zip(quarters_Q2, values_Q2)), columns=['Quarters', 'Number of Vehicles'])
df_battery_electric_Q3 = pd.DataFrame(list(zip(quarters_Q3, values_Q3)), columns=['Quarters', 'Number of Vehicles'])
df_battery_electric_Q4 = pd.DataFrame(list(zip(quarters_Q4, values_Q4)), columns=['Quarters', 'Number of Vehicles'])

# Change ',' sign into '.'
df_battery_electric['Number of Vehicles'] = df_battery_electric['Number of Vehicles'].str.replace(',', '.').astype('float64')
df_battery_electric_Q1['Number of Vehicles'] = df_battery_electric_Q1['Number of Vehicles'].str.replace(',', '.').astype('float64')
df_battery_electric_Q2['Number of Vehicles'] = df_battery_electric_Q2['Number of Vehicles'].str.replace(',', '.').astype('float64')
df_battery_electric_Q3['Number of Vehicles'] = df_battery_electric_Q3['Number of Vehicles'].str.replace(',', '.').astype('float64')
df_battery_electric_Q4['Number of Vehicles'] = df_battery_electric_Q4['Number of Vehicles'].str.replace(',', '.').astype('float64')

#Check the data frame
pd.DataFrame(df_battery_electric_Q4)
```

```python
# Merge the two dataframes on the quarter column
merged_df = pd.merge(df_battery_electric, df_quarterly, left_on='Quarters', right_index=True)
merged_df_Q1 = pd.merge(df_battery_electric_Q1, df_quarterly, left_on='Quarters', right_index=True)
merged_df_Q2 = pd.merge(df_battery_electric_Q2, df_quarterly, left_on='Quarters', right_index=True)
merged_df_Q3 = pd.merge(df_battery_electric_Q3, df_quarterly, left_on='Quarters', right_index=True)
merged_df_Q4 = pd.merge(df_battery_electric_Q4, df_quarterly, left_on='Quarters', right_index=True)

pd.DataFrame(merged_df)
```

```python
correlation = merged_df['Number of Vehicles'].corr(merged_df['NOx'])
correlation_Q1 = merged_df_Q1['Number of Vehicles'].corr(merged_df_Q1['NOx'])
correlation_Q2 = merged_df_Q2['Number of Vehicles'].corr(merged_df_Q2['NOx'])
correlation_Q3 = merged_df_Q3['Number of Vehicles'].corr(merged_df_Q3['NOx'])
correlation_Q4 = merged_df_Q4['Number of Vehicles'].corr(merged_df_Q4['NOx'])

print("Correlation between Number of Vehicles and NOx:", correlation)
print("Correlation between Number of Vehicles and NOx on 1st quarter:", correlation_Q1)
print("Correlation between Number of Vehicles and NOx on 2nd quarter:", correlation_Q2)
print("Correlation between Number of Vehicles and NOx on 3rd quarter:", correlation_Q3)
print("Correlation between Number of Vehicles and NOx on 4th quarter:", correlation_Q4)
```

The negative correlation coefficient of -0.5358 suggests that there is a moderate negative relationship between the Number of Vehicles and NOx (nitrogen oxides) emissions. This means that as the number of vehicles increases, the NOx emissions decrease and vice versa. However, it is important to note that correlation does not imply causation, and there may be other factors that influence NOx emissions.

```python
from scipy.stats import linregress

x = merged_df['Number of Vehicles']
y = merged_df['NOx']

slope, intercept, r_value, p_value, std_err = linregress(x, y)
print("Slope:", slope)
print("Intercept:", intercept)
print("R-squared:", r_value**2)
```

Slope: The slope of the line (also known as the regression coefficient) represents the rate of change of NOx with respect to the number of vehicles. A negative slope of -0.02898 suggests that as the number of vehicles increases by one unit, the NOx decreases by approximately 0.029 units.

Intercept: The intercept represents the value of NOx when the number of vehicles is zero. In this case, the intercept is 106.913, which means that if there were no vehicles, the NOx would still be at a level of approximately 107 units.

R-squared: The R-squared value measures how well the regression line fits the data. It represents the proportion of the variation in the dependent variable (NOx) that is explained by the independent variable (number of vehicles). An R-squared value of 0.2871 indicates that approximately 28.71% of the variation in NOx can be explained by the number of vehicles. The remaining 71.29% of the variation is attributed to other factors not included in the model.

```python
from sklearn.linear_model import LinearRegression

# create scatter plot
plt.scatter(merged_df_Q1['Number of Vehicles'], merged_df_Q1['NOx'], c='blue')
plt.scatter(merged_df_Q2['Number of Vehicles'], merged_df_Q2['NOx'], c='orange')
plt.scatter(merged_df_Q3['Number of Vehicles'], merged_df_Q3['NOx'], c='green')
plt.scatter(merged_df_Q4['Number of Vehicles'], merged_df_Q4['NOx'], c='purple')

# fit linear regression
X = merged_df[['Number of Vehicles']]
y = merged_df['NOx']
model = LinearRegression().fit(X, y)
r_sq = model.score(X, y)
slope = model.coef_[0]
intercept = model.intercept_

# create line plot
x_range = np.linspace(X.min(), X.max(), num=100)
y_range = slope * x_range + intercept
plt.plot(x_range, y_range, color='red')

# add labels and title
plt.xlabel('Number of EV Vehicles')
plt.ylabel('NOx')
plt.title('Scatter plot with regression line (R-squared = {:.3f})'.format(r_sq))

# show plot
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
# extract variables
x = merged_df["Number of Vehicles"].values.reshape(-1, 1)
y = merged_df["NOx"].values.reshape(-1, 1)

list_r2 = []

# Find the best degree of freedom
for i in range(1,  len(x)):
    poly = PolynomialFeatures(degree=i)
    x_poly = poly.fit_transform(x)
    poly_reg = LinearRegression()
    poly_reg.fit(x_poly, y)

    # predict y values using fitted model
    y_pred = poly_reg.predict(x_poly)

    # calculate R-squared
    ss_res = np.sum((y - y_pred)**2)
    ss_tot = np.sum((y - np.mean(y))**2)
    r_squared = 1 - (ss_res / ss_tot)
    list_r2.append(r_squared)

max_r_squared = max(list_r2)
best_degree = list_r2.index(max_r_squared) + 1
print("Maximum R-squared:", max_r_squared)
print("Best degree of freedom:", best_degree)
```
